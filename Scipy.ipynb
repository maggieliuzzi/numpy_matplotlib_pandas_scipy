**PDF and CDF**

```
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm  # standard normal distribution  # for others: beta, gamma
```

```
x = np.linspace(-6, 6, 1000)
```

PDF

```
# PDF of x values under standard normal distribution
fx = norm.pdf(x, loc=0, scale=1)  # for normal distribution, loc=mean, scale=std
```

```
# Plotting PDF
plt.plot(x, fx);  # if 1 arg: y axis, if 2: x, y
```

CDF

```
# CDF of standard normal
Fx = norm.cdf(x, loc=0, scale=1)
```

```
plt.plot(x, Fx);
```

Log PDF

```
# Log PDF of the standard normal
logfx = norm.logpdf(x, loc=0, scale=1)
```

```
plt.plot(x, logfx);
```

Applying convolution to an image to create a blurrying filter using a Gaussian filter

```
im = Image.open("cute.jpg")
```

```
gray = np.mean(im, axis=2)  # converting image to grayscale, 2D
```

```
x = np.linspace(-6, 6, 50)
fx = norm.pdf(x, loc=0, scale=1)  # PDF for these points

filt = np.outer(fx, fx)  # 2D Gaussian filter (2D spherical Gaussian with constants that we'll ignore): outer product of f(x) with itself

plt.imshow(filt, cmap="gray");
```

```
from scipy.signal import convolve2d
```

```
out = convolve2d(gray, filt)
```

```
plt.subplot(1,2,1)  # n rows, n cols, position # 1x2 grid of plots
plt.imshow(gray, cmap="gray")
plt.subplot(1,2,2)
plt.imshow(out, cmap="gray");
```
