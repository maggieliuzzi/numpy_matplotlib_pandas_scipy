```
import numpy as np
import matplotlib.pyplot as plt
```

**Line Charts**

```
x = np.linspace(0, 20, 1000)  # 1D array with 1000 evenly-spaced points between 0 and 20
y = np.sin(x) + 0.2 * x
```

```
plt.plot(x, y)
plt.xlabel("input")
plt.ylabel("output")
plt.title("My Plot");  # semicolons avoid printing the last return
```

```
plt.show()  # if not using Jupyter notebook
```

**Scatter Plots**

```
X = np.random.randn(100, 2)  # 100 observations, 2 dimensions, drawn from normal distribution, so centered at 0
```

```
plt.scatter(X[:,0], X[:,1])  # scatter(horiz axis, vertical axis)  # first and second cols of X
```

```
X = np.random.randn(200, 2)
# centering first half of points at 3, 3
X[:50] += 3  # rows 0-50  # in-place
Y = np.zeros(200)  # array of zeros of size 200
Y[:50] = 1  # first half
plt.scatter(X[:,0], X[:,1] , c=Y);  # c: colour, 1D array of ints
```

**Histograms**
> For plotting distributions



```
X = np.random.randn(10000)  # centered around 0, and approx 95% between -+1.96
plt.hist(X)
```

```
plt.hist(X, bins=50);   # centered around 0, and approx 95% between -+1.96
```

```
X = np.random.random(10000)  # samples from the uniform 0-1 distribution
plt.hist(X);  # plt.hist(X, bins=50);
```

```
im = Image.open("landscape.jpeg")  # PIL PngImageFile
```

```
arr = np.array(im)  # images are represented as arrays, H x W x Colour (RGB channels)
'''  arr[0]
array([[234, 217, 207],
       [234, 217, 207],
       [234, 217, 207],
       ...,
       [224, 215, 208],
       [224, 215, 208],
       [224, 215, 208]], dtype=uint8)  # 8-bit unsigned intigers (0-255, since 2^8 = 256)
'''
# arr[0][0]  # array([234, 217, 207], dtype=uint8)  # same as doing arr[0,0]
arr  # size: 2001000  # (667, 1000, 3)
```

```
plt.imshow(arr)  # plt.imshow(im)
```

```
gray = arr.mean(axis=2)  # shape: (667, 1000)  collapsed colour dimension
plt.imshow(gray)  # heatmap
```

```
plt.imshow(gray, cmap="gray")
```

```
# Not grayscale
mountain = Image.open("black-and-white.jpeg")
arr_moun = np.array(mountain)
arr_moun.shape  # (312, 500, 3)
arr_moun[0][0]  # array([40, 40, 40], dtype=uint8)
plt.imshow(arr_moun)
```

```
arr_moun_gray = arr_moun.mean(axis=2)  # shape: (312, 500)
plt.imshow(arr_moun_gray, cmap="gray")
```

```
gray_flower = Image.open("grayscale.jpeg")
arr_flower = np.array(gray_flower)
arr_flower.shape  # (183, 275, 3)
arr_flower[0][0]  # array([103, 103, 103], dtype=uint8)
plt.imshow(arr_flower)
```

```
arr_flower_gray = arr_flower.mean(axis=2)  # shape: (183, 275)
plt.imshow(arr_flower, cmap="gray")
```
